### Podział zbioru

```python
from fast_ml.model_development import train_valid_test_split

X = dane.drop(['class'], axis=1)
y = dane['class']

X_train, y_train, X_valid, y_valid, X_test, y_test = train_valid_test_split(dane, target = 'class',
                                                                            train_size=0.7, valid_size=0.2, test_size=0.1)
print(X_train.shape), print(y_train.shape)
print(X_valid.shape), print(y_valid.shape)
print(X_test.shape), print(y_test.shape)
```

### Wybór modelu i uczenie

#### Regresja logistyczna

```python
t0=time.time()
lr = LogisticRegression(max_iter=100,random_state=0)
lr.fit(X_train, y_train) # estymacja modelu

pred_train1=lr.predict(X_train) # predykcja na danych treningowych
pred_test1=lr.predict(X_test) # predykcja na danych testowych
pred_valid1=lr.predict(X_valid) # predykcja na danych walidacyjnych

# Ocena jakości modelu 
score=accuracy_score(y_train,pred_train1)
print("Accuracy Score dla danych treningowych:", score) 
score1=accuracy_score(y_test,pred_test1)
print("Accuracy Score dla danych testowych:", score1) 
score2=accuracy_score(y_valid,pred_valid1)
print("Accuracy Score dla danych walidacyjnych:", score2) 

#print("Accuracy on test set: {:.3f}".format(knn.score(X_test_std, y_test)))

print('Czas na trening modelu:' , time.time()-t0) 
```

### Selekcja zmiennych

Test chi-squared mierzy, czy istnieje statystycznie istotny związek między zmiennymi kategorycznymi input i output.
W analizie dokonanej przy zmiennej binarnej wyjściowej.

```python
cat_df = dane.select_dtypes(include='object')
cat_df['class'] = dane['class']

x_cat = dane.drop(['class'], axis=1)
y_cat = cat_df['class']
          
oe = OrdinalEncoder()
x_cat_transform = oe.fit_transform(x_cat)

cfs = SelectKBest(score_func=f_classif, k=10)
cfs.fit(x_cat_transform, y_cat)

cat_feature_score = pd.DataFrame({'Column':x_cat.columns,'Score':cfs.scores_})
print(cat_feature_score.nlargest(5,'Score'))

#Select features
cat_selected = [i for i in cat_feature_score.nlargest(5,'Score').Column]


#Plot the scores
plt.figure(figsize=(14,10))
sns.barplot(x='Score',y='Column',data=cat_feature_score.sort_values('Score',ascending=False), color='#F72585')
plt.show()
```
